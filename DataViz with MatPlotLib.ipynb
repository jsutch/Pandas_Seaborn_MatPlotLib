```python
fig = plt.figure()
```

```python
fig = plt.figure()
# left, bottom, width and height
# [.1,.1,.8,.8]
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'ro')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Outer Rectangle - red dots')
axes2.plot(y,x,'g^')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Inner Rectangle - green triangles')
```

## Figure Size, Aspect Ratio and DPI

**with figure:**

```python
# the figsize tuple is the width and height of the figure in Inches
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# the figsize tuple is the width and height of the figure in Inches
fig = plt.figure(figsize=(8,1),dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

**with subplots:**

```python
# the figsize tuple is the width and height of the figure in Inches
fig,ax = plt.subplots(figsize=(7.5,.75),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# the figsize tuple is the width and height of the figure in Inches
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# Fix overlap
# the figsize tuple is the width and height of the figure in Inches
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Add a Legend

**using loc settings:**

```python
# setting x, y and title labels
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Add labels to the plot for the legend to reference
ax.plot(x,x**2,'g-',label='x squared')
ax.plot(x,x**3,'r-',label='x cubed')
# set the location. 0-10. 'Best'-'Center'
# Empty is loc=0 e.g. "best"
# loc 2 is upper left (the same as 'best', for this case)
ax.legend(loc=2) 
```

```python
# setting x, y and title labels
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Add labels to the plot for the legend to reference
ax.plot(x,x**2,'g-',label='x squared')
ax.plot(x,x**3,'r-',label='x cubed')
# set the location. 0-10. 'Best'-'Center'
# Empty is loc=0 e.g. "best"
# loc 10 is center
ax.legend(loc=10) 
```

**using tuple coordinates:**

```python
# setting x, y and title labels
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Add labels to the plot for the legend to reference
ax.plot(x,x**2,'g-',label='x squared')
ax.plot(x,x**3,'r-',label='x cubed')
# Using tuples to indicate coordinate placement
ax.legend(loc=(0.1,0.4)) 
```

# Setting line colors, widths and options

#### color

```python
# Base plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# can accept common colors like 'green' or 'orange'
# or can accept RGP hex codes - orange in hex
ax.plot(x,y,color='#FF8C00')
```

#### linewidth

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# linewidth can also be abbreviated with: lw=
ax.plot(x,y,color='#FF8C00',linewidth=4)
```

#### set alpha

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# can accept common colors like 'green' or 'orange'
# or can accept RGP hex codes - orange in hex
ax.plot(x,y,color='#FF8C00',linewidth=6,alpha=0.25)
```

#### linestyle

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# can abbreviate linestyle to ls=
ax.plot(x,x**2,color='#FF8C00',lw=4,linestyle='-')
ax.plot(x,x**2.5,color='green',lw=4,linestyle='--')
ax.plot(x,x**3,color='red',lw=2,linestyle='-.')
ax.plot(x,x**3.2,color='purple',lw=2,linestyle=':')
ax.plot(x,x**3.5,color='red',lw=2,linestyle='steps')
```

#### line markers

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# can abbreviate linestyle to ls=
ax.plot(x,x**2,color='#FF8C00',lw=2,linestyle='-',marker='o',markersize=10)
ax.plot(x,x**2.5,color='green',lw=2,linestyle='--',marker='s',markersize=7)
ax.plot(x,x**3,color='red',lw=2,linestyle='-.',marker='P',markersize=8)
ax.plot(x,x**3.2,color='purple',lw=2,linestyle=':',marker='*',markersize=10)
ax.plot(x,x**3.5,color='red',lw=2,linestyle='steps',marker='>',markersize=8)
ax.plot(x,x**3.6,color='indigo',lw=1,linestyle='--',marker='1',markersize=20)
```

#### adding definition to the markers

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='blue',lw=2,linestyle='-.',marker='>',markersize=20,markerfacecolor='orange',markeredgewidth=1)
ax.plot(x,x**3,color='green',lw=3,linestyle=':',marker='s',markersize=20,markerfacecolor='pink',markeredgewidth=5,markeredgecolor='red')
ax.plot(x,x**3.6,color='indigo',lw=3,linestyle='-.',marker='p',markersize=20,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='cyan')
```

#### Control Over Axis Appearance

```python
# Add color parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color='blue',lw=2,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
