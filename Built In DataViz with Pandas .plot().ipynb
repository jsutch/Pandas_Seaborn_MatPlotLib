```python
import numpy as np
import pandas as pd
import seaborn as sns
# matplotlib config
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["patch.force_edgecolor"] = True # in matplotlib, edge borders are turned off by default. 
sns.set_style("darkgrid") # set a grey grid as a background
```

```python
# Time Series
df1 = pd.read_csv('df1',index_col=0)
```

```python
# Index is a Time Series
df1.head()
```

```python
# Non Time Series
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

### Histogram

```python
# automatically create a histogram 
# (calls matplotlib)
df1['A'].hist()
```

```python
# All plots can be done with a command off the dataframe, and params within plot
# Make a histogram
df1['A'].plot(kind='hist',bins=30)
```

```python
# instead of calling kind, call .plot.kind()
# e.g. .plot.hist()
df1['A'].plot.hist(bins=30)
```

```python
df2.head()
```

### Area Plot

```python
# Calling an Area plot. Adjust transparency with alpha=
df2.plot.area(alpha=0.4)
```

### Bar Plot

```python
# Take the index value as category
df2.plot.bar()
```

```python
# Create a Stacked bar
df2.plot.bar(stacked=True)
```

```python
# Back to histogram
df1['A'].plot.hist(bins=50)
```

```python
# Line Plots
df1.plot.line()
```

```python
# Make this more coherent 
# Original lecture used "df1.plot.line(x=df1.index,y='B')", but the syntax changed in pandas
df1['B'].plot.line(figsize=(12,3),lw=1)
```

```python
# Scatterplots
df1.plot.scatter(x='A',y='B')
```

```python
# set the color from another column with c='columnname'
df1.plot.scatter(x='A',y='B',c='C')
```

```python
# add a colormap
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# Setting Size with s= from the C column
df1.plot.scatter(x='A',y='B',s=df1['C'])
```

```python
# Adjusting size of the datapoints
df1.plot.scatter(x='A',y='B',s=df1['C']*50)
```

```python
# adding color and palette back in based on D
df1.plot.scatter(x='A',y='B',s=df1['C']*50,c='D',cmap='coolwarm')
```

```python
# BoxPlots
df2.plot.box()
```

```python
# Bivariate Data - Hexagonal Bin Plot
df = pd.DataFrame(np.random.randn(1000,2),columns=['A','B'])
```

```python
df.head()
```

```python
# like a scatterplot but with Hex bins
df.plot.hexbin(x='A',y='B')
```

```python
# enlarge the size of the Hexagons for clarity
df.plot.hexbin(x='A',y='B',gridsize=25)
```

```python
# add cmap
df.plot.hexbin(x='A',y='B',gridsize=25,cmap='coolwarm')
```

```python
# Kernel Density Plot
df2['a'].plot.kde()
```

```python
# also .density()
df2['a'].plot.density()
```

```python
# Use multiple columns for comparison
df2.plot.density()
```
